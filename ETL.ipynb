```python
# Normalizamos primero las comillas para la columna belongs_to_collection:

import json

def fix_quotes(x):
    if pd.isnull(x):
        return x
    try:
        return json.loads(x.replace("'", '"'))
    except json.JSONDecodeError:
        return {}

moviesdf['belongs_to_collection'] = moviesdf['belongs_to_collection'].apply(fix_quotes)
```

```python
# Este ciclo for imprime los nombres de las colecciones si existen y None si no hay una colección asociada a la película.

for collection in moviesdf['belongs_to_collection']:
    if collection is not None:
        print(collection.get('name'))
```

```python
# Miramos el dataframe y observaos que la columna genres tiene datos tipo diccionario, que debenos transformar
moviesdf.head()
```

```python
# Convertios las comillas simples en comillas dobles en la columna 'genres' del dataframe 'moviesdf', utilizando la biblioteca json. Luego aplicamos 
# la función 'get_genre_name' a la columna 'genres' para obtener el nombre del género en formato de cadena de texto separado por "|" en la nueva 
# columna 'genre_names'.

import json

def fix_quotes(x):
    if pd.isnull(x):
        return x
    try:
        return json.loads(x.replace("'", '"'))
    except json.JSONDecodeError:
        return []
        
moviesdf['genres'] = moviesdf['genres'].apply(fix_quotes)

def get_genre_name(genre_list):
    names = [genre['name'] for genre in genre_list]
    return '|'.join(names) if names else None

moviesdf['genre_names'] = moviesdf['genres'].apply(get_genre_name)
```

```python
moviesdf.head()
```

```python
# Con este ciclo se recorre cada fila de la columna "genre_names", luego para cada fila se valida si el valor en la columna "genre_names" 
# es diferente de None. Si el valor no es nulo, se divide el valor en una lista de géneros utilizando el carácter '|' como separador de elementos 
# y se guarda como la variable "genre_list". 

for genres in moviesdf['genre_names']:
    if genres is not None:
        genre_list = genres.split('|')
        for genre in genre_list:
            print(genre)
    else:
        print('No genres')
```

```python
# Eliminamos la columna original
moviesdf.drop(columns=['genres'], inplace=True)
```

```python
# Miramos los cambios realizados
moviesdf
```

```python
# Se transforma la columna spoken_languages que contienen datos en formato json, luego definimos una función parse_json que acepta una cadena json (x), 
# intentamos parsearla a un objeto Jjson, para que devuelva el objeto si tiene éxito; si no, devuelve un diccionario vacío ({}). Después, la columna 
# "spoken_languages" se convierte utilizando esta función, es decir que se reemplazan los valores presentes en dicha columna que son 
# cadenas json con los respectivos objetos json producidos por parse_json.

def parse_json(x):
    try:
        return json.loads(x.replace("'", "\""))
    except:
        return {}

moviesdf["spoken_languages"] = moviesdf["spoken_languages"].apply(parse_json)

# Definimos una segunda función get_spoken_languages que espera un objeto json (x). Si el objeto es un diccionario vacío, devuelve la cadena "No languages", 
# de lo contrario, combina los nombres de idiomas contenidos en este objeto json en una sola cadena separada por "|" y devuelve esta cadena.

def get_spoken_languages(x):
    if x == {}:
        return "No languages"
    else:
        return "|".join([d["name"] for d in x])

moviesdf["spoken_languages"] = moviesdf["spoken_languages"].apply(get_spoken_languages)
print(moviesdf["spoken_languages"])
```

```python
# Funcion para obtener el nombre de los objetos anidados

def get_object_names(x):
    try:
        object_list = eval(x)
        object_names = [object_dict['name'] for object_dict in object_list]
        return ', '.join(object_names)
    except:
        return ''
```

```python
# Creamos una nueva columna como nombre de los paises productores  
moviesdf['production_countries_names'] = moviesdf['production_countries'].apply(get_object_names)
```

```python
# Creamos una nueva columna, como nombre produccion companias
moviesdf['production_companies_names'] = moviesdf['production_companies'].apply(get_object_names)
```
